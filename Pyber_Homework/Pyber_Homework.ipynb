## Total Fares by City Type

```python
city_type_pyber = complete_dataset_df.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
city_type_pyber
```

```python
# Calculate Type Percents
fares_city_type_pie = city_type_pyber.sum()['fare']
fares_city_type_pie
```

```python
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0,0, 0.4)

#Create the pie chart.
plt.pie(fares_city_type_pie, labels = fares_city_type_pie.index, autopct = "%1.2f%%", colors = colors, explode = explode, shadow = True, startangle = 150)

plt.axis("equal")

# Legend.
plt.legend(loc = "upper right")

# Save Figure
plt.savefig('fares_city_type_pie.png')

plt.show()
plt.show()
```
